```python
import os
import glob
import pprint

from ruamel.yaml import YAML

YAML = YAML()
```

```python
fnames = sorted(glob.glob('../artifacts/*/*.yml'))

data = {
    'cython': 0,
    'pybind11': 0,
    'total': 0,
    'matplotlib': 0,
    'numpy': 0,
    'pandas': 0,
    'scipy': 0,
    'pytest': 0,
    'mock': 0,
    'unittest': 0,
    'typing': 0,
    'nose': 0,
    'future': 0,
    'pinning_exact': 0,
    'pinning_equiv': 0,
    'pinning_range': 0,
    'pinning_lower': 0,
    'pinning_upper': 0,
    'total_missing': 0,
    'requirement': 0,
    'total_pkgs': 0,
}
ignore_keys = [
    'total', 
    'pinning_exact', 
    'pinning_equiv',
    'pinning_range', 
    'pinning_lower',
    'pinning_upper',
    'total_missing',
    'requirement',
    'total_pkgs',
]
problems = set()

for fname in fnames:
    with open(fname, 'r') as fp:
        res = YAML.load(fp)
    
    if 'No broken' in res['pip_check']:
        continue
    else:
        data['total_pkgs'] += 1
        
    print("pkg:", fname.split('/')[-2])
    for line in res['pip_check'].split('\n'):
        if line in problems:
            continue

        if len(line) == 0:
            continue
        
        problems.add(line)
        print("   ", line)
        
        
        if 'has requirement' in line:
            data['requirement'] += 1
        
        data['total'] += 1
        if '==' in line:
            data['pinning_exact'] += 1
        elif '~=' in line:
            data['pinning_equiv'] += 1
        elif '<' in line and '>' in line:
            data['pinning_range'] += 1
        elif '<' in line and '>=' in line:
            data['pinning_range'] += 1
        elif '<=' in line and '>' in line:
            data['pinning_range'] += 1
        elif '<=' in line and '>=' in line:
            data['pinning_range'] += 1
        elif '>=' in line or '>' in line:
            data['pinning_lower'] += 1
        elif '<=' in line or '<' in line:
            data['pinning_upper'] += 1
        
        if 'which is not installed' in line:
            data['total_missing'] += 1
        
        for key in data:
            if key not in ignore_keys:
                if key in line and 'which is not installed' in line:
                    data[key] += 1        
```

```python
print(pprint.pformat(data))
```

```python
assert data['total_missing'] + sum(v for k, v in data.items() if 'pinning_' in k) == data['total']
```

```python
print('total conda packages pip checked:', len(fnames))
print('    frac of conda packages flagged:', data['total_pkgs'] / len (fnames))
print(' ')
print('fractions of lines output by pip check\n'
      '  (can have more than one line per package):')
print('    missing python package:', data['total_missing'] / data['total'])
print('    wrong ver python package:', 1.0 - data['total_missing'] / data['total'])
print('        exact      :', data['pinning_exact'] / data['total'])
print('        range/equiv:', (data['pinning_range'] + data['pinning_equiv']) / data['total'])
print('        lower      :', data['pinning_lower'] / data['total'])
print('        upper      :', data['pinning_upper'] / data['total'])

print(' ')
print('specific python packages (missing only, total %d):' % data['total_missing'])
for key in data:
    if key not in ignore_keys:
        print('    %s:' % key, data[key])
```
